### Dados:
<ul style='font-size: 14px; line-height: 2; text-align: justify;'>
<li><b>school</b> - escola do aluno (binário: 'GP' - Gabriel Pereira ou 'MS' - Mousinho da Silveira)</li>
<li><b>sex</b> - sexo do aluno (binário: 'F' - feminino ou 'M' - masculino)</li>
<li><b>age</b> - idade do aluno (numérico: de 15 a 22 anos)</li>
<li><b>address</b> - tipo de endereço da casa do aluno (binário: 'U' - urbano ou 'R' - rural)</li>
<li><b>famsize</b> - tamanho da família (binário: 'LE3' - menor ou igual a 3 ou 'GT3' - maior que 3)</li>
<li><b>Pstatus</b> - status de coabitação dos pais (binário: 'T' - vivendo juntos ou 'A' - separados)</li>
<li><b>Medu</b> - educação da mãe (numérico: 0 - nenhuma, 1 - ensino fundamental (4ª série), 2 - 5ª a 9ª série, 3 - ensino médio ou 4 - ensino superior)</li>
<li><b>Fedu</b> - educação do pai (numérico: 0 - nenhuma, 1 - ensino fundamental (4ª série), 2 - 5ª a 9ª série, 3 - ensino médio ou 4 - ensino superior)</li>
<li><b>Mjob</b> - profissão da mãe (nominal: 'professora', relacionada à 'saúde', 'serviços' civis (por exemplo, administrativos ou policiais), 'em_casa' ou 'outro')</li>
<li><b>Fjob</b> - profissão do pai (nominal: 'professor', relacionada à 'saúde', 'serviços' civis (por exemplo, administrativos ou policiais), 'em_casa' ou 'outro')</li>
<li><b>reason</b> - motivo para escolher esta escol (nominal: 'perto de 'casa', 'reputação' da escola, preferência de 'curso' ou 'outro')</li>
<li><b>guardian</b> - responsável pelo aluno (nominal: 'mãe', 'pai' ou 'outro')</li>
<li><b>traveltime</b> - tempo de viagem de casa para a escola (numérico: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. a 1 hora, ou 4 - >1 hora)</li>
<li><b>studytime</b> - tempo semanal de estudo (numérico: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas, ou 4 - >10 horas)</li>
<li><b>failures</b> - número de reprovações passadas (numérico: n se 1<=n<3, senão 4)</li>
<li><b>schoolsup</b> - suporte educacional extra (binário: sim ou não)</li>
<li><b>famsup</b> - suporte educacional da família (binário: sim ou não)</li>
<li><b>paid</b> - aulas extras pagas no assunto do curso (Matemática ou Português) (binário: sim ou não)</li>
<li><b>activities</b> - atividades extracurriculares (binário: sim ou não)</li>
<li><b>nursery</b> - frequentou a escola maternal (binário: sim ou não)</li>
<li><b>higher</b> - deseja cursar ensino superior (binário: sim ou não)</li>
<li><b>internet</b> - acesso à Internet em casa (binário: sim ou não)</li>
<li><b>romantic</b> - em um relacionamento romântico (binário: sim ou não)</li>
<li><b>famrel</b> - qualidade dos relacionamentos familiares (numérico: de 1 - muito ruim a 5 - excelente)</li>
<li><b>freetime</b> - tempo livre após a escola (numérico: de 1 - muito baixo a 5 - muito alto)</li>
<li><b>goout</b> - sair com amigos (numérico: de 1 - muito baixo a 5 - muito alto)</li>
<li><b>Dalc</b> - consumo de álcool em dias úteis (numérico: de 1 - muito baixo a 5 - muito alto)</li>
<li><b>Walc</b> - consumo de álcool no fim de semana (numérico: de 1 - muito baixo a 5 - muito alto)</li>
<li><b>health</b> - estado de saúde atual (numérico: de 1 - muito ruim a 5 - muito bom)</li>
<li><b>absences</b> - número de faltas na escola (numérico: de 0 a 93)</li>
<li><b>G1</b> - nota do primeiro período (numérico: de 0 a 20)</li>
<li><b>G2</b> - nota do segundo período (numérico: de 0 a 20)</li>
<li><b>G3</b> - nota final (numérico: de 0 a 20, alvo de saída)</li>

</ul>
